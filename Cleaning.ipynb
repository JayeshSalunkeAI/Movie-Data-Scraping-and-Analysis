```python
import pandas as pd
import re
```

```python
movies_df = pd.read_csv("movies_data.csv")
```

```python
movies_df.head()
```

```python
movies_df.info()
```

**RunTime Cleaning**

```python
ragex_runtime = r'\d*'
movies_df["RunTime"] = movies_df["RunTime"].apply(lambda x:re.findall(ragex_runtime, x)[0] if x != None else None)
movies_df.head(10)
```

```python
movies_df['numeric_col'] = pd.to_numeric(movies_df['RunTime'], errors='coerce')
```

```python
movies_df['RunTime'] = movies_df['numeric_col'].copy()
movies_df
```

```python
movies_df = movies_df.drop(columns=['numeric_col'])
```

**Rating Cleaning**

```python
ragex_rating = r'\d.\d*'
movies_df["Rating"] = movies_df["Rating"].apply(lambda x:re.findall(ragex_rating, x)[0] if x != None else None)
movies_df.head(10)
```

```python
movies_df['Rating'] = pd.to_numeric(movies_df['Rating'], errors='coerce')
```

```python
movies_df.info()
```

```python
movies_df = movies_df.dropna()
```

```python
movies_df.info()
```

```python
movies_df.to_csv("Clean_movie_data.csv", index=False)
```
